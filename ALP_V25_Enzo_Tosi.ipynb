Peça a quantidade de horas trabalhadas e o valor por hora. Se trabalhou até 40 horas, calcule o salário normalmente (horas × valor por hora). Se trabalhou mais de 40 horas, as horas extras devem ter um adicional de 50% (horas extras × valor por hora × 1.5). Crie um algoritmo em Python que realize esse cálculo corretamente.

```python
horas_trabalhadas = float(input("Digite a quantidade de horas trabalhadas: "))
valor_hora = float(input("Digite o valor por hora: "))

if horas_trabalhadas <= 40:
    salario = horas_trabalhadas * valor_hora
else:
    horas_extras = horas_trabalhadas - 40
    salario = (40 * valor_hora) + (horas_extras * valor_hora * 1.5)

print(f"O salário total é R$ {salario:.2f}")
```

Crie um algoritmo que faça uma calculadora, que repite 0 para sair:
Deve mostrar o seguinte menu na tela:
Calculadora V25
Digite:

1 - Multiplos de 5

2 - Números Pares

3 - Números Prismos

4 - Conversor de temperatura

0 - sair

---------------------------
Cada vez que o usuário digitar um número deve entrar na função apedir os dados ao usuário e efutuar o cálculo.
Mostrar a resposta e solicitar novamente o que ele quer fazer:

```python
while True:
    print("\n===========================")
    print("   Calculadora V25")
    print("   Digite:")
    print("   1 - Múltiplos de 5")
    print("   2 - Números Pares")
    print("   3 - Números Primos")
    print("   4 - Conversor de temperatura")
    print("   0 - Sair")
    print("---------------------------")

    opcao = input("Escolha uma opção: ")


    if opcao == '0':
        print("Saindo da calculadora. Até mais!")
        break


    elif opcao == '1':
        num = int(input("Digite um número para verificar múltiplos de 5: "))
        if num % 5 == 0:
            print(f"O número {num} é múltiplo de 5.")
        else:
            print(f"O número {num} não é múltiplo de 5.")


    elif opcao == '2':
        num = int(input("Digite um número para verificar se é par: "))
        if num % 2 == 0:
            print(f"O número {num} é par.")
        else:
            print(f"O número {num} é ímpar.")


    elif opcao == '3':
        num = int(input("Digite um número para verificar se é primo: "))

        if num < 2:
            print(f"O número {num} não é primo.")
        else:
            is_primo = True
            for i in range(2, int(num**0.5) + 1):
                if num % i == 0:
                    is_primo = False
                    break

            if is_primo:
                print(f"O número {num} é primo.")
            else:
                print(f"O número {num} não é primo.")


    elif opcao == '4':
        print("\n--- Conversor de Temperatura ---")
        print("1 - Celsius para Fahrenheit")
        print("2 - Fahrenheit para Celsius")
        opcao_temp = input("Escolha uma opção (1 ou 2): ")

        if opcao_temp == '1':
            celsius = float(input("Digite a temperatura em Celsius: "))
            fahrenheit = (celsius * 9/5) + 32
            print(f"{celsius}°C é igual a {fahrenheit}°F")

        elif opcao_temp == '2':
            fahrenheit = float(input("Digite a temperatura em Fahrenheit: "))
            celsius = (fahrenheit - 32) * 5/9
            print(f"{fahrenheit}°F é igual a {celsius}°C")

        else:
            print("Opção inválida para o conversor de temperatura.")


    else:
        print("Opção inválida. Por favor, digite um número de 0 a 4.")
```
